```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
a= sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',data=a)
```

```python
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

```python
b=sns.load_dataset('tips')
```

```python
print(b)
```

```python
sns.relplot(x='day',y='tip',kind='line',data=b)
```

```python
sns.catplot(x='day',y='tip',data=b)
```

```python
sns.boxplot(x='day',y='total_bill',data=b)
```

```python
sns.set(style="white", color_codes=False)
sns.boxplot(x='day',y='total_bill',data=b)
```

```python
sns.despine(offset=10, trim=True)
```
